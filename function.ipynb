```python
# write a menu driven program
# addition press + from add
# substraction press from subtract
# product press * from multiply
# divide press / from divide

print("+ , - , * , /")
choice = input("enter your choice :")
num1= int(input("enter your num1 "))
num2= int(input("enter your num2"))
def addition (num1 , num2):
    add = num1+num2
    return add

if choice == "+":
    res = addition(num1 , num2)
    print(res)

def minus (num1 , num2):
    add = num1-num2
    return add

if choice == "-":
    res = minus(num1 , num2)
    print(res)

def multiply (num1 , num2):
    add = num1*num2
    return add

if choice == "*":
    res = multiply(num1 , num2)
    print(res)

def division (num1 , num2):
    add = num1 / num2
    return add

if choice == "/":
    res = division(num1 , num2)
    print(res)

```

```python
# wap to check to count vowels and consonantss
text = input("enter your string ")
def check(text):
    cons = 0
    vowels = 0

    for ch in text:
        if ch in "aeiouAEIOU":
            vowels += 1
        else:
            cons += 1

    return cons, vowels

cons, vowels = check(text)

print("Consonants =", cons)
print("Vowels =", vowels)


    

```
