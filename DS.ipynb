```python
#write a program to print only vowels
emp_name = ["aman", "shivam", "rohan", "anuj"]
vowels = "aeiouAEIOU"
vow = ""

for name in emp_name:
    for ch in name:
        if ch in vowels:
            vow += ch

print(vow)
```

```python
marks = [90,55,33,22,78,99,23,33]

marks_tag = []

for i in marks:
    if i >= 55:
        marks_tag.append(str(i) + ":pass")
    elif i < 22:
        marks_tag.append(str(i) + ":fail")

print(marks_tag)
```

```python
# waf to generate only odd and store in the list using compr.
def odd(A,B):
    odd_num = []
    for i in range(A,B+1):
        if i%2 !=0:
            odd_num.append(i)
    return odd_num
print(odd(10,19))
```

```python
marks = [90,55,33,22,78,99,23,33]
marks_odd = []
marks_even = []
for i in marks:
    if i % 2 == 0:
        marks_even.append(i)
    elif i % 2 != 0:
        marks_odd.append(i)
        
print(marks_odd)
print(marks_even)
```
