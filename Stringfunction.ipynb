```python
# Data Masking


num = "987654321660"

for i in range(len(num)):
    if i < len(num) - 4:
        print("X", end="")
    else:
        print(num[i], end="")
```

```python
def mask_number(data, visible=4):
    masked_part = 'X' * (len(data) - visible)
    visible_part = data[-visible:]
    return masked_part + visible_part

print(mask_number('987654321660'))
```

```python
num = "9876543210"

print("X" * (len(num) - 4) + num[-4:])
```

```python
data_list = ['this', 'is', 'python']
vowels = "AEIOUaeiou"

new_list = []

for word in data_list:
    new_word = ""
    for ch in word:
        if ch in vowels:
            new_word += "X"
        else:
            new_word += ch
    new_list.append(new_word)

print(new_list)
```
