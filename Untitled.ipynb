```python
# 4. Write a Python program to check if a number provided by the user is prime or not.
num = int(input("Enter the number: "))

if num <= 1:
    print("Not a prime number")
else:
    i = 2
    is_prime = True

    while i < num:
        if num % i == 0:
            is_prime = False
            break
        i += 1

    if is_prime:
        print("Prime number")
    else:
        print("Not a prime number")
```

```python
# Write a Python program to generate the Fibonacci sequence up to a specified 
# number of terms. 

num = int(input("Enter number of terms: "))

a = 0
b = 1
i = 1

while i <= num:
    print(a)
    c = a + b
    a = b
    b = c
    i += 1
```

```python
# 4. Write a Python program to check whether a string entered by the user is a 
# palindrome. 
str = input("Enter a string")
rev = ""
i= len(str)-1
while i >= 0:
    rev = rev + str[i]
    i-=1
print(rev)
if rev == str:
    print("Given string is Palindrome")
else:
    print("Given String is not Palindrom")
```
