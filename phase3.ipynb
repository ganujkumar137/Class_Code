Advance operation with business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [restaurent_id ,2021,2022,2023,2024]
sales_data=np.array([
    [1,15000,100,25000,26000],
    [2,15000,20000,25000,26000],
    [3,15000,20000,25000,26000],
    [4,15000,20000,25000,26000],
    [5,15000,20000,25000,26000],

])

print("=== Zomoto Sales Analys ===")
print(sales_data.shape)
print("sample data first 3 restuarent ", sales_data[:,1:])
print(np.sum(sales_data,axis=0))
yearly = np.sum(sales_data[:,1:],axis=1)
print(yearly)
print(np.sum(yearly))
```

Minimum sales per restreaunt

```python
min_sale = np.min(sales_data[:,1:],axis=1)
print(min_sale)

max_sale = np.max(sales_data[:,1:],axis=0)
print(max_sale)
```

Average sales per restaurent

```python

avg_sale =np.average(sales_data[:,1:],axis=1)
print(avg_sale)

cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumsum Sale per Year across all restraunt")
plt.xlabel("Years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
